#### Step 1 : importing the libraries

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

#### Step 2: initializing the config file and the frozen model

```python
config_file = "C:\\Users\\admin\\Desktop\\stage\\The sparks foundation\\Object-Detection-main\\Object-Detection-main\\ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
frozen_model = "C:\\Users\\admin\\Desktop\\stage\\The sparks foundation\\Object-Detection-main\\Object-Detection-main\\frozen_inference_graph.pb"
```

#### Step 3 : initializing the pre-trained model 

```python
model= cv2.dnn_DetectionModel(frozen_model ,config_file )
```

```python
classLabels = []
file_name = 'C:\\Users\\admin\\Desktop\\stage\\The sparks foundation\\Object-Detection-main\\Object-Detection-main\\Labels.txt'
with open(file_name, 'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean(127.5)
model.setInputSwapRB(True)
```

#### Step 4 : loading and plotting the image

```python
img =cv2.imread('C:\\Users\\admin\\Desktop\\stage\\The sparks foundation\\Object-Detection-Optical-Character-Recognition--main\\Object-Detection-Optical-Character-Recognition--main\\train_data_set\\joaquin phoenix\\images_021.jpg')
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img , cv2.COLOR_BGR2RGB))
```

```python
classIndex , confidence , bbox =model.detect(img,confThreshold=0.5)
```

#### Step 5 : drawing the image results 

```python
print(classIndex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf , boxes in zip (classIndex.flatten(), confidence.flatten(), bbox):
     cv2.rectangle(img,boxes,(0,0,255),2 )
     cv2.putText(img,classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40),font ,fontScale=font_scale,color=(0,255,0),thickness=3)                              
```

```python
plt.imshow(img)
```

#### Step 6 :loading the video and drawing the results


```python
cap =cv2.VideoCapture("C:\\Users\\admin\\Desktop\\stage\\The sparks foundation\\Object-Detection-main\Object-Detection-main\\Road_traffic_video2.mp4")

if not cap.isOpened():
    cap= cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("cannot open video")

font_scale = 3 
font= cv2.FONT_HERSHEY_PLAIN

while True:
    ret,frame = cap.read()
    ClassIndex,confidence, bbox =model.detect(frame,confThreshold=0.55)
    print(ClassIndex)
    if(len(ClassIndex)!=0):
        for ClassInd , conf ,boxes in zip(ClassIndex.flatten(), confidence.flatten() ,bbox) :
                if(ClassInd<=80):
                    cv2.rectangle(frame,boxes,(0,255,0),2 )
                    cv2.putText(frame,classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font ,fontScale=font_scale,color=(255,0,0), thickness=3)

    cv2.imshow('object Detction',frame)
    if cv2.waitKey(2) & 0xFF == ord('q'):
       break
cap.release()
cv2.destroyAllWindows()
```
